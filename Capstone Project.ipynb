# Segmentation and Clustering Neighborhoods in Toronto

```python
!pip install lxml
!pip install beautifulsoup4
import pandas as pd # For Analysis
import numpy as np # Used to Handle data
import random # For generating random numbers
import requests 
from geopy.geocoders import Nominatim # Conversion latitude and longitude values
from IPython.display import Image # Generating Images
from IPython.core.display import HTML 
from IPython.display import display_html
from pandas.io.json import json_normalize
!conda install -c conda-forge folium=0.5.0 --yes
import folium # plotting library
from bs4 import BeautifulSoup
from sklearn.cluster import KMeans
import matplotlib.cm as cm
import matplotlib.colors as colors
```

```python
# Scrapping the wikipedia page
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(source,'lxml')
from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```

```python
# Table is converted to Pandas Frame
dfs = pd.read_html(tab)
df=dfs[0]
df.head()
```

```python
# Data Preprocess and Data Cleaning
df1 = df[df.Borough != 'Not assigned'] # Let's drop the rows where borough is not assigned
df2 = df1.groupby(['Postal Code','Borough'], sort=False).agg(', '.join) # Combining where there is a match in postal code
df2.reset_index(inplace=True)
# Replace neighborhoods not assigned by borough
df2['Neighborhood'] = np.where(df2['Neighborhood'] == 'Not assigned',df2['Borough'], df2['Neighborhood'])
df2
```

```python
df2.shape # Data frame shape
```

```python
#Importing csv file which has longitude and latitude
lat_lon = pd.read_csv('https://cocl.us/Geospatial_data')
lat_lon.head()
```

```python
# Getting Latitude and Longitude
lat_lon.rename(columns={'Postcode':'Postal Code'},inplace=True)
df3 = pd.merge(df2,lat_lon,on='Postal Code')
df3.head()
```

```python
# Getting all the rows which has Toronto in their Borough
df4 = df3[df3['Borough'].str.contains('Toronto',regex=False)]
df4
```

```python
# Visualizing the dataframe - Folium
map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)
for lat,lng,borough,neighbourhood in zip(df4['Latitude'],df4['Longitude'],df4['Borough'],df4['Neighborhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```

```python
# Using K means (Clustring of neighborhoods)
k=5
toronto_clustering = df4.drop(['Postal Code','Borough','Neighborhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
df4.insert(0, 'Cluster Labels', kmeans.labels_)
df4
```

```python
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=10) # Creting the map
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]
markers_colors = [] # Adding Markers to the map
for lat, lon, neighbourhood, cluster in zip(df4['Latitude'], df4['Longitude'], df4['Neighborhood'], df4['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
